```python
import cv2
import matplotlib.pyplot as plt


# Створемо клас для обробки зображення
class ImageProcessor:
    def __init__(self, image_path):
        self.image = cv2.imread(image_path) # Зчитування зображення з диску

    # Функція для відображення зображення    
    def show_image(self):
        plt.imshow(cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

    # Функція для збереження зображення на диску
    def save_image(self, save_path):
        cv2.imwrite(save_path, self.image)
    
    # Функція для зміни контрастності зображення (задання коефіцієнту контрастності та яскравості) 
    def change_contrast_brightness(self, contrast, brightness):
        adjusted_image = cv2.convertScaleAbs(self.image, alpha=contrast, beta=brightness)
        self.image = adjusted_image
    
    # Функція для зміни яскравості зображення
    def change_brightness(self, brightness):
        self.change_contrast_brightness(1.0, brightness)
    
    # Функція для зміни насиченості зображення
    def change_saturation(self, saturation):
        hsv_image = cv2.cvtColor(self.image, cv2.COLOR_BGR2HSV)
        hsv_image[:, :, 1] = hsv_image[:, :, 1] * saturation
        self.image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)

    # Функція для зміни розміру зображення
    def resize_image(self, new_width, new_height):
        self.image = cv2.resize(self.image, (new_width, new_height))
    
    # Функція для обрізання зображення
    def crop_image(self, start_x, start_y, end_x, end_y):
        self.image = self.image[start_y:end_y, start_x:end_x]
    
    # Функція для повороту зображення на заданий кут
    def rotate_image(self, angle):
        height, width = self.image.shape[:2]
        rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1.0)
        self.image = cv2.warpAffine(self.image, rotation_matrix, (width, height))
```

```python
# Приклад використання
image_path = input("Введіть шлях до зображення: ")
processor = ImageProcessor(image_path)
processor.show_image()
```

```python
# Зміна контрастності і яскравості
contrast = float(input("Введіть коефіцієнт контрастності: "))
brightness = int(input("Введіть коефіцієнт яскравості: "))
processor.change_contrast_brightness(contrast, brightness)
processor.show_image()
```

```python
# Зміна яскравості
brightness = int(input("Введіть коефіцієнт яскравості: "))
processor.change_brightness(brightness)
processor.show_image()
```

```python
# Зміна насиченості
saturation = float(input("Введіть коефіцієнт насиченості: "))
processor.change_saturation(saturation)
processor.show_image()
```

```python
# Зміна розміру зображення
new_width = int(input("Введіть нову ширину зображення: "))
new_height = int(input("Введіть нову висоту зображення: "))
processor.resize_image(new_width, new_height)
processor.show_image()
```

```python
# Обрізання зображення
start_x = int(input("Введіть початкову координату X для обрізання: "))
start_y = int(input("Введіть початкову координату Y для обрізання: "))
end_x = int(input("Введіть кінцеву координату X для обрізанння: "))
end_y = int(input("Введіть кінцеву координату Y для обрізанння: "))
processor.crop_image(start_x, start_y, end_x, end_y)
processor.show_image()
```

```python
# Поворот зображення
angle = float(input("Введіть кут повороту зображення в градусах: "))
processor.rotate_image(angle)
processor.show_image()
```

```python
# Збереження зображення на диск
save_path = input("Введіть шлях для збереження зображення: ")
processor.save_image(save_path)
```
